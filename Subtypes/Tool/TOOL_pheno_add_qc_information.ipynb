# Add QC information
The purpose of this notebook is to take an existing phenotype file and add QC information

```python
import os
import re
import pandas as pd
```

```python
# Paths
pheno_path = '/data1/abide/Pheno/pheno.csv'
qc_motion_path = '/data1/abide/Pheno/motion_all.csv'
merged_path = '/data1/abide/Pheno/merged_pheno.csv'
```

```python
pheno = pd.read_csv(pheno_path)
qc_motion = pd.read_csv(qc_motion_path)
# Rename header of QC file
colnames = qc_motion.columns.values
colnames[0] = 'SUB_ID'
qc_motion.columns = colnames
# Change subject IDs
qc_names = [int(re.search('(?<=X)[0-9]*',row['SUB_ID']).group()) for index, row in qc_motion.iterrows()]
qc_motion['SUB_ID'] = qc_names
```

```python
qc_motion.head()
```

```python
pheno.head()
```

```python
combined_pheno = pheno.merge(qc_motion, on='SUB_ID')
```

```python
combined_pheno.head()
```

```python
out_pheno = combined_pheno.loc[combined_pheno.duplicated('SUB_ID').values==False]
```

```python
out_pheno.to_csv(merged_path, index=False)
```
